```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Dileep Sathya\OneDrive\Desktop\MLOPS\first_end_to_end_ML\artifacts\data_transformation\data\transformed_data.csv")

df.head()
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.2)
print(train.shape)
print(test.shape)
```

```python
train.head()
```

```python
x_train=train.drop(["loan_status"],axis=1)
x_test=test.drop(["loan_status"],axis=1)
from sklearn.preprocessing import LabelEncoder

# Initialize a nested dictionary to store mappings for all columns
mappings = {}

# Identify categorical columns
categorical_list = [col for col in x_train.columns if x_train[col].dtype == 'object']

# Encode each categorical column
for col in categorical_list:
    le = LabelEncoder()  # Create a new LabelEncoder for each column
    x_train[col] = le.fit_transform(x_train[col])
    x_test[col] = le.transform(x_test[col])
    
    # Store both mappings (forward in the nested dictionary
    mappings[col] = {v:int(k) for k, v in zip(le.transform(le.classes_), le.classes_)}


# Print the nested dictionary with mappings for each column
print(mappings)

# Display the first 10 rows of the encoded training data
x_train.head(10)
```
